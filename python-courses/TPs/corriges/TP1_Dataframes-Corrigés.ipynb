```python
#remplacer les valeurs manquantes par les éléments les plus fréquents
salaries_df['salary_currency']=salaries_df['salary_currency'].fillna("USD")
```

```python
# Trouver un moyen intelligent pour remplacer les valeurs null dans ces colonnes 
salaries_df['employment_type'].value_counts()
```

```python
salaries_df['employment_type']=salaries_df['employment_type'].fillna("FT")
```

```python
salaries_df["company_location"] = salaries_df['company_location'].fillna('US')
```
