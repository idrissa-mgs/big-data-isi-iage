On a une lists de livres, chaque livre étant représenté par un dictionaires


1 - Parcourir la liste et afficher les différents titre de livre (boucle for, print)

2 - Créer 3 nouvelles listes à partir de cette lists ces nouvelles listes contiendront respectivement les différents auteurs, année de publication et genres (boucle for ou list compression, append)

```python
books = [
    {
        'title': 'The Great Gatsby',
        'author': 'F. Scott Fitzgerald',
        'year_published': 1925,
        'genre': 'Fiction'
    },
    {
        'title': 'To Kill a Mockingbird',
        'author': 'Harper Lee',
        'year_published': 1960,
        'genre': 'Fiction'
    },
    {
        'title': '1984',
        'author': 'George Orwell',
        'year_published': 1949,
        'genre': 'Science Fiction'
    },
    {
        'title': 'The Hobbit',
        'author': 'J.R.R. Tolkien',
        'year_published': 1937,
        'genre': 'Fantasy'
    }
]
```

```python
books[0].get("title")
```

```python
# qestion 1
for b in books:
    print(b['title'])
```

```python
# question 2
authors_list = []
years_published_list = []
genres_list = []
```

```python
authors_list = [b['author'] for b in books]
years_published_list = [b["year_published"] for b in books]
genres_list = [b["genre"] for b in books]
```

```python
print(authors_list)
print(years_published_list)
print(genres_list)
```

##### Les fonctions
Une fonction est un bloc de code qui effectue une tâche spécifique et peut être appelé depuis d'autres parties du programme. Les fonctions permettent de diviser les programmes volumineux en morceaux plus petits

```python
def trigram(x: str):
    return x[:3].upper()
```

```python
def prenom_nom(pre: str, nom: str):
    return pre + " " + nom
```
